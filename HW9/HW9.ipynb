# Problem 1

```python
# import Glop package
from ortools.linear_solver import pywraplp as glp
import lptools as lpt
```

```python
# each row (sublist) corresponds to a single student
# each element in a row (sublist) indicates the student's preference for a project

preference = [[2,5,3,2,4,4,5],
              [1,5,1,2,2,4,5],
              [2,5,4,4,5,5,3],
              [2,5,3,1,5,4,4],
              [2,4,3,1,5,5,5],
              [5,5,2,2,2,3,5],
              [2,5,2,3,4,5,5],
              [2,5,3,1,3,5,4],
              [1,5,2,3,4,5,1],
              [3,5,4,2,3,5,5],
              [2,5,2,2,2,2,5],
              [1,5,1,2,2,5,4],
              [1,4,1,2,1,4,5],
              [1,5,1,2,2,5,4],
              [1,5,1,1,4,5,4],
              [2,5,3,2,4,5,3],
              [1,5,1,2,4,3,4],
              [3,4,2,1,3,5,4],
              [5,4,3,5,5,4,5],
              [5,5,1,4,5,5,1],
              [2,5,1,1,4,5,4],
              [1,4,4,5,2,2,3],
              [4,5,1,1,5,5,5],
              [1,4,1,3,5,2,3],
              [1,3,1,2,2,4,5],
              [2,5,4,4,5,5,3],
              [2,4,3,2,3,5,4],
              [3,5,3,1,4,5,5]]
```

```python
# Assign each student to an id, will be labeled 0-26

agents = {}

for i in range(len(preference)):
    student_preference = preference[i]
    dict_index = str(i)
    agents[dict_index] = (1, student_preference)

print(agents)
```

```python
# Create tasks (projects) 1-7, will have id's 0-6 respectively
# Each project has 4 members

tasks = {}
num_members = 4
for i in range(7):
    dict_index = str(i)
    tasks[dict_index] = num_members

print(tasks)
```

```python
#Create LP model object
mymodel = glp.Solver('Students',glp.Solver.GLOP_LINEAR_PROGRAMMING)
inf = mymodel.infinity()
mymodel.Objective().SetMaximization()
```

```python
# create all assignment variables and a dictionary of dictionaries to access them
assign =  dict()
for a in agents:
    assign[a] =  dict()
    (b,coeff_lst) =  agents[a]
    for (t,c) in zip(tasks,coeff_lst):
        assign[a][t] =  mymodel.NumVar(0, inf, a + "-" + t)
        mymodel.Objective().SetCoefficient(assign[a][t],c)
```

```python
print(assign)
```

```python
# create task constraints
for t in tasks:
    lb = ub = tasks[t]
    
    constr = mymodel.Constraint(lb,ub,t)
    for a in agents:
        constr.SetCoefficient(assign[a][t],1)
```

```python
lpt.print_model(mymodel)
```

```python
# create agent constraints
for a in agents:
    (b,coeff_lst) = agents[a] 
    constr = mymodel.Constraint(-inf , b, a)
    for t in tasks:
        constr.SetCoefficient(assign[a][t], 1)
```

```python
lpt.print_model(mymodel)
```

```python
#solve model and display results
status = mymodel.Solve()
print('Solution Status =',status)
print('Optimal Value = %.2f' % mymodel.Objective().Value())
for v in mymodel.variables():
    print('%s = %.2f' % (v.name(),v.solution_value()))
```

```python
# display all variable information
print('Variable    LB   Value    UB   Reduced Cost')
for v in mymodel.variables():
    print('%8s  %5.1f  %5.1f  %5.1f  %5.2f' % (v.name(),v.lb(),v.solution_value(),v.ub(),v.reduced_cost()))
```

```python
# only print nonzero assignment variables
print('Variable    LB   Value    UB   Reduced Cost')
for v in mymodel.variables():
    if v.solution_value() != 0:
        print('%8s  %5.1f  %5.1f  %5.1f  %5.2f' % (v.name(),v.lb(),v.solution_value(),v.ub(),v.reduced_cost()))
```

# Problem 2

```python
#Create LP model object
mymodel = glp.Solver('Utilities',glp.Solver.GLOP_LINEAR_PROGRAMMING)
inf = mymodel.infinity()
mymodel.Objective().SetMinimization()
```

```python
inf = mymodel.infinity()

         # name: supply capacity, objective coefficient list
sources = {'Southern': (500 , [10,20,15 ]),  
           'Northwest': (400 , [12,15,18 ])
          } 

        # name: demand quantity required
demands = {'Hamilton': 400, 'Butler': 200, 'Clermont': 300}  

```

```python
# create all shipment variables and a dictionary of dictionaries to access them
ship = dict()
for s in sources:
    ship[s] = dict()
    (b,coeff_lst) = sources[s]
    for (d,c) in zip(demands,coeff_lst):
        ship[s][d] = mymodel.NumVar(0, inf, s + '_' + d)
        mymodel.Objective().SetCoefficient(ship[s][d] ,c )
```

```python
print(ship)
```

```python
# create demand constraints
for d in demands:
    lb = ub = demands[d]
    constr = mymodel.Constraint(lb ,ub ,d )
    for s in sources:
        constr.SetCoefficient(ship[s][d] , 1)
```

```python
# create source/supply constraints
for s in sources:
    (b,coeff_lst) = sources[s]
    constr = mymodel.Constraint(-inf ,b ,s )
    for d in demands:
        constr.SetCoefficient(ship[s][d] ,1 )
```

```python
lpt.print_model(mymodel)
```

```python
#solve model and display results
status = mymodel.Solve()
print('Solution Status =',status)
print('Optimal Value = %.2f' % mymodel.Objective().Value())
for v in mymodel.variables():
    print('%s = %.2f' % (v.name(),v.solution_value()))
```

```python
# display all variable information
print('Variable    LB   Value    UB   Reduced Cost')
for v in mymodel.variables():
    print('%8s  %5.1f  %5.1f  %5.1f  %5.2f' % (v.name(),v.lb(),v.solution_value(),v.ub(),v.reduced_cost()))
```

```python
# only print nonzero shipment variables
print('Variable    LB   Value    UB   Reduced Cost')
for v in mymodel.variables():
    if v.solution_value() != 0:
        print('%8s  %5.1f  %5.1f  %5.1f  %5.2f' % (v.name(),v.lb(),v.solution_value(),v.ub(),v.reduced_cost()))
```

```python
#display constraint information
print('Constraint    LB    Value  UB     Dual')
for (c,lhs) in zip(mymodel.constraints(),mymodel.ComputeConstraintActivities()):
    print('%10s  %5.1f  %5.1f  %5.1f  %5.2f' % (c.name(),c.lb(),lhs,c.ub(),c.dual_value()))
```
