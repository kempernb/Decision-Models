# Fort Thomas Brewing Co. - Sim 2

```python
import random as rd
import statistics as st
import numpy.random as npr
import matplotlib.pyplot as plt
```

```python
#input parameters
price = 18.00       # $ per case
cost = 12.00        # $ per case
salvage = 8.00      # $ per case
mean_dmd = 1000.0   # cases
```

```python
#simulation parameter analysis for the production quantity from 950 to 1050

trials = 1000
prod_qty = list(range(950, 1050))
mean_profit = list()
min_profit = list()
max_profit = list()
range_profit = list()
for order in prod_qty:
    sample = FTBC_sim(order,trials)
    mean_profit.append(st.mean(sample))
    max_profit.append(max(sample))
    min_profit.append(min(sample))
    range_profit.append(max(sample) - min(sample))
    
    
```

```python
# user function to generate simulation sample

def FTBC_sim(order,trials):
    sample = list()
    for i in range(trials):
        demand = npr.poisson(mean_dmd)
        if demand < order:
            qty_sold = demand
        else: 
            qty_sold = order
        qty_left = order - qty_sold
        sales_revenue = price * qty_sold
        salvage_revenue = salvage * qty_left
        total_cost = cost * order
        profit = sales_revenue + salvage_revenue - total_cost
        sample.append(profit)
    return(sample)
```

```python
# mean profit
plt.plot(prod_qty,mean_profit)
```

```python
# min profit
plt.plot(prod_qty,min_profit)
```

```python
# max profit
plt.plot(prod_qty,max_profit)
```

```python
# range profit
plt.plot(prod_qty,range_profit)
```

```python
order = prod_qty[mean_profit.index(max(mean_profit))]
print('The production quantity which maximizes mean profit is %d cases.' % order)
print('The expected profit is $%7.2f.' % (max(mean_profit)))
```

```python
# perfect information simulation

trials = 1000
sample = list()
for i in range(trials):
    demand = npr.poisson(mean_dmd)
    order = demand
    if demand < order:
        qty_sold = demand
    else: 
        qty_sold = order
    qty_left = order - qty_sold
    sales_revenue = price * qty_sold
    salvage_revenue = salvage * qty_left
    total_cost = cost * order
    profit = sales_revenue + salvage_revenue - total_cost
    sample.append(profit)
sample_mean = st.mean(sample)

    
evpi = sample_mean - max(mean_profit)

print('The mean profit for the perfect information case is $%5.2f.' % sample_mean)
print('The expected value of perfect information is $%5.2f.' % evpi)
```
